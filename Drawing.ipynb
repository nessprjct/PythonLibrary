```python
from google.colab.patches import cv2_imshow
import cv2

def to_int(data):
      if type(data)==list or type(data)==tuple:
          new_data=[int(datum) for datum in data]
          return new_data

def draw_point(image,color,point:tuple,radius):
      radius = radius
      color = color
      thickness = 1
      # point=self.to_int(point)
      point=point
      image = cv2.circle(image, point, radius, color, thickness)
      return image

def draw_box(image,start_point,end_point):
      color = (0, 225, 0)
      thickness = 2
      image = cv2.rectangle(image, start_point, end_point, color, thickness)
      return image

def draw_line(image,start_point, end_point):
      color = (0, 255, 0)
      thickness = 2
      image = cv2.line(image, start_point, end_point, color, thickness)
      return image

def add_text(image,text,point):
        #point : coordinates where to put the text
        font = cv2.FONT_HERSHEY_SIMPLEX
        fontScale = 0.5
        color = (0, 0, 0)
        thickness = 1
        image = cv2.putText(image, text, point, font,
                          fontScale, color, thickness, cv2.LINE_AA)
        return image

#------------------------

path = r'/content/IMG_20221210_091102.jpg'
image = cv2.imread(path)

color = (255, 0, 0)
point = (515, 120)
font = cv2.FONT_HERSHEY_DUPLEX
thickness = 2

cv2.ellipse(image, point, (32,60),40,10,40, color, thickness)
#triangle rec
# pts = np.array([[5, 25], [25, 5], [5, 5]])
# pointt = np.array([505, 140])
# ptts = pts + pointt
# cv2.polylines(image, [ptts], True, (255,0,0), 2)


image = draw_line(image, (900, 145), (100, 145))
image = draw_line(image, (510, 145), (510, 700))

image = add_text(image, '162', (573, 230))
image = add_text(image, '62', (530, 325))
image = add_text(image, '-76', (450, 325))
image = add_text(image, '169', (400, 230))

image = cv2.putText(image,'86', (525,170), font, 0.5, (255,0,0), 1)

#------------------------

image1 = draw_box(image, (510, 145), (435, 306))
image2 = draw_box(image, (510, 302), (570, 145))


a1 = (435 - 510) * (306 - 145)
a2 = (570 - 510) * (145 - 302)

if a1 < a2:

    image = add_text(image, 'right higher', (620, 230))
    cv2.arrowedLine(image, (610, 340), (610, 145), (0, 0, 255), 2, 8, 0, 0.1)

    image = add_text(image, 'right closer', (530, 360))
    cv2.arrowedLine(image, (610, 340), (510, 340), (0, 0, 255), 2, 8, 0, 0.1)

else:

    image = add_text(image, 'left higher', (305, 230))
    cv2.arrowedLine(image, (395,340), (395,145), (0,0,255), (2), 8,0,0.1)

    image = add_text(image, 'left closer', (403, 360))
    cv2.arrowedLine(image, (395,340), (510,340), (0,0,255), (2), 8,0,0.1)

cv2_imshow(image)
```
